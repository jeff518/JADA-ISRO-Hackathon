```python
# utils/predictor.py

import joblib

model_45m = joblib.load("../models/xgboost_45m_model.pkl")
model_6h = joblib.load("../models/xgboost_6h_model.pkl")
model_12h = joblib.load("../models/xgboost_12h_model.pkl")

def predict_all(features):

    return {
        "45m": float(model_45m.predict(features)[0]),
        "6h": float(model_6h.predict(features)[0]),
        "12h": float(model_12h.predict(features)[0]),
    }
```
